```python
def zscore_normalise_features(x):
    mu=np.mean(x,axis=0)
    sd=np.std(x,axis=0)
    x_norm=(x-mu)/sd
    return x_norm
```

```python
wdf_num=zscore_normalise_features(wdf_num)
w_num=wdf_num.to_numpy()
```

```python
def kMeans_init_centroids(X, K):
    randidx = np.random.permutation(X.shape[0])
    centroids = X[randidx[:K]]
    return centroids
centroids=kMeans_init_centroids(w_num,4)
```

```python
def find_closest_centroids(X, centroids):
    K = centroids.shape[0]
    idx = np.zeros(X.shape[0], dtype=int)
    for i in range(X.shape[0]):
        l=X[i]-centroids[0]
        l**=2
        d=np.sum(l)
        min_d=d
        index=0
        
        for j in range(1,K):
            l=X[i]-centroids[j]
            l**=2
            d=np.sum(l)
            if (d<min_d):
                min_d=d
                index=j
        idx[i]=index
    return idx
```

```python
def compute_centroids(X, idx, K):
    m, n = X.shape
    centroids = np.zeros((K, n))
    for i in range(K):
        total=np.zeros((1,n))
        count=0
        for j in range (m):
            if idx[j]==i:
                total+=X[j]
                count+=1
        total/=count
        centroids[i]=total
    return centroids   
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components = 2)
W2D = pca.fit_transform(w_num)
pca.explained_variance_ratio_
```

```python
def run_kMeans(X, K, max_iters=10, plot_progress=False):
    randidx = np.random.permutation(X.shape[0])
    centroids = X[randidx[:K]]
    print(centroids)
    m, n = X.shape
    for i in range(max_iters):
        print("K-Means iteration %d/%d" % (i, max_iters-1))
        idx = find_closest_centroids(X, centroids)
        centroids = compute_centroids(X, idx, K)
    plt.show() 
    return centroids, idx
centroids,idx=run_kMeans(W2D,4)
```
