```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('train_logistic.csv')
is_male=pd.get_dummies(df['Sex'],drop_first=True)
df['is_male']=is_male
df["is_male"] = df["is_male"].astype(int)
df.drop(['Name','Sex'],axis=1,inplace=True)
x_train=df[["Pclass","Age","Siblings/Spouses Aboard","Parents/Children Aboard","Fare",'is_male']]
y_train=df["Survived"]
x_train=x_train.to_numpy()
y_train=y_train.to_numpy()
m,n=np.shape(x_train)
w=np.zeros((n,))
b=0
print(x_train)
```

```python
def z_score_normalise(x):
    mu=np.mean(x,axis=0)
    sigma=np.std(x,axis=0)
    x_norm=(x-mu)/sigma
    return x_norm,mu,sigma
x_train_norm,mu,sigma=z_score_normalise(x_train)
```

```python
def sigmoid(x,w,b):
    m,n=np.shape(x)
    z=np.zeros((m,))
    for i in range(m):
        z[i]=np.dot(x[i],w)+b
    m=np.shape(z)[0]
    g_wb=np.ones((m,))
    g_wb*=(1/(1+np.exp(-z)))
    return g_wb
```

```python
def compute_gradient_logistic(x,y,w,b):
    m,n=np.shape(x)
    g_wb=sigmoid(x,w,b)
    dj_dw=np.zeros((n,))
    dj_db=0
    err=g_wb-y
    dj_db=np.sum(err)
    for i in range(m):
        for j in range(n):
            dj_dw[j]+=err[i]*x[i,j]
    dj_dw/=m
    dj_db/=m
    print(dj_dw)
    print(dj_db)
    return dj_dw,dj_db
```

```python
def gradient_descent(x,y,w,b,num_iters,alpha):
    j_history=[]
    for i in range(num_iters):
        dj_dw,dj_db=compute_gradient_logistic(x,y,w,b)
        w-=alpha*dj_dw
        b-=alpha*dj_db
        j_history.append(compute_gradient_logistic(x,y,w,b))
    return w,b,j_history   
```

```python
iters=1000
alpha=0.1
w_fin,b_fin,j_hist=gradient_descent(x_train_norm,y_train,w,b,iters,alpha)
```

```python
df_test=pd.read_csv("test_logistic.csv")
is_male=pd.get_dummies(df_test['Sex'],drop_first=True)
df_test['is_male']=is_male
df["is_male"] = df["is_male"].astype(int)
df_test.drop(['Name','Sex'],axis=1,inplace=True)
x_test=df_test[["Pclass","Age","Siblings/Spouses Aboard","Parents/Children Aboard","Fare",'is_male']]
m,n=np.shape(x_test)
y_test=df_test[['Survived']]
x_test=x_test.to_numpy()
y_test=y_test.to_numpy().reshape((m,))
m,n=np.shape(x_test)
x_test_norm=(x_test-mu)/sigma
y_pred=np.zeros((m,))
g_wb=sigmoid(x_test_norm,w_fin,b_fin)
for i in range(m):
    if (g_wb[i]>=0.5):
        y_pred[i]=1
    else:
        y_pred[i]=0
y_pred=y_pred.astype(int)
print(y_test)
print(y_pred)
count=0
for i in range (m):
    if y_test[i]==y_pred[i]:
        count+=1
print(count)
```
